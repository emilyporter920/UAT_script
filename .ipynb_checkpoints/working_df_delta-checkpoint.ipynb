```python
# Read in CSVs
df_old = pd.read_csv('Test_Data/df_old.csv')
df_new = pd.read_csv('Test_Data/df_new.csv')
```

```python
df_old = pd.DataFrame(df_old)
df_new = pd.DataFrame(df_new)
```

```python
df_new = df_new.reset_index()
df_new.head()
```

```python
df_diff3 = pd.merge(df_old, df_new, how='outer', indicator='Exist')
df_diff3.head()
```

```python
# This takes the differences and puts them into a dataframe (there is the same amount of left_only as there is right_only because they are the differences)
df_diff3 = df_diff3.query('Exist != "both"')
df_diff3
```

```python
# This is what is in the new dataframe but not in the old dataframe
right_only = df_diff3.query('Exist == "right_only"')

right_only.columns = right_only.columns.str.upper()

right_only
```

```python
# This is what is in the old dataframe but not in the new dataframe
left_only = df_diff3.query('Exist == "left_only"')

left_only.columns = left_only.columns.str.upper()

left_only
```

```python
# This one shows what appears inside of the UAT that does not appear in the GILBDW

result1 = right_only[right_only.apply(tuple, 1).isin(left_only.apply(tuple, 1))]

# These can be new row entries or changes to existing rows.
result2 = left_only[~left_only.apply(tuple, 1).isin(right_only.apply(tuple, 1))]

df_diff = pd.DataFrame(result2)

print('----------------------------------------------------------------------------------------------------------------')

print('This shows the changes. These are located inside of NEW dataframe that have either been changed or are new rows.')

print('----------------------------------------------------------------------------------------------------------------')

df_diff
```

```python
# This one shows what appears inside of the UAT that does not appear in the GILBDW

result1 = right_only[~right_only.apply(tuple, 1).isin(left_only.apply(tuple, 1))]

df_diff1 = pd.DataFrame(result1)

print('----------------------------------------------------------------------------------------------------------------')

print('This shows the changes. These are located inside of OLD dataframe that have either been changed or are new rows.')

print('----------------------------------------------------------------------------------------------------------------')

df_diff1
```

```python
# Append the two dataframes using this method: https://www.geeksforgeeks.org/python-pandas-dataframe-append/
# DataFrame.append(other, ignore_index=False, verify_integrity=False, sort=None) 

df_diff = df_diff.append(df_diff1, ignore_index=True)
```
